````python
import tiktoken
from dotenv.ipython import load_dotenv
import os
````

````python
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)
````

````python
prompt = "you are an expert analyst in feeling analysis"
````

````python
tokens =  tokennizer.encode(prompt)
````

````python
print(len(tokens))
````

````python
for token in tokens:
    t  = tokennizer.decode_single_token_bytes(token).decode("utf-8")
    print(t, end="|")
````

````python
def tokens_count(prompt:str, model:str="o200k_base"):
    tokennizer = tiktoken.get_encoding(model)
    return len(tokennizer.encode(prompt))
````

````python
print(tokens_count(prompt="you are the best"))
````

````python
from langchain_core.tools import tool
````

````python
@tool
def free_image_api(prompt: str) -> str:
    return f"https://api.stability.ai/v1/generation/{hash(prompt)}/image.png"
````

````python
llm5 = ChatGroq(model="llama-3.3-70b-versatile")
llm_with_tools_2 = llm5.bind_tools([free_image_api])
````

````python
system_message = """
You are an expert in the sentiment analysis domain.
Perform an aspect-based sentiment analysis of product reviews.
Each review may contain one or more of the following aspects: screen, keyboard, pad.
For each review provided as input:
- Identify the presence of at least one of the three aspects (screen, keyboard, pad)
- Assign a sentiment polarity (positive, negative, or neutral) to each aspect
Organize your response as a JSON object with the following keys:
  - category:[list of aspects]
  - polarity:[list of corresponding polarities for each aspect]
If one of the aspects is not present in the user's review, assume it is neutral.
"""
````
