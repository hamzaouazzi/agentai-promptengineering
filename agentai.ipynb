```python
import tiktoken
from dotenv.ipython import load_dotenv
import os
```

```python
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)
```

```python
prompt = "you are an expert analyst in feeling analysis"
```

```python
tokens =  tokennizer.encode(prompt)
```

```python
print(len(tokens))
```

```python
for token in tokens:
    t  = tokennizer.decode_single_token_bytes(token).decode("utf-8")
    print(t, end="|")
```

```python
def tokens_count(prompt:str, model:str="o200k_base"):
    tokennizer = tiktoken.get_encoding(model)
    return len(tokennizer.encode(prompt))
```

```python
print(tokens_count(prompt="you are the best"))
```

```python
llm4 = ChatOpenAI(model="gpt-5.2")
llm_with_tools = llm4.bind_tools([
    {"type":"image_generation","quality":"high"}
])
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
```

```python
resp = llm_with_tools.invoke(input=[
    SystemMessage(""),
    HumanMessage("I want a picture of cat coding python")
])
```

```python
from IPython.display import Image
import base64
```

```python
Image(base64.b64decode(resp.content_blocks[0]['base64']))
```

```python
from langchain_core.tools import tool
```

```python
@tool
def free_image_api(prompt: str) -> str:
    """Génère une image gratuite et retourne l'URL."""
    return f"https://api.stability.ai/v1/generation/{hash(prompt)}/image.png"
```

```python
llm5 = ChatGroq(model="llama-3.3-70b-versatile")
llm_with_tools_2 = llm5.bind_tools([free_image_api])
```

```python
messages = [
    SystemMessage(""),
    HumanMessage("I want a picture of a cat coding Python")
]
```

```python
resp = llm_with_tools_2.invoke(messages)
print(resp)
```
